### Q.1 Delete the elements in an linked list whose sum is equal to zero

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    def delete_elements_with_zero_sum(self):
        current = self.head
        while current:
            sum_value = 0
            temp = current
            while temp:
                sum_value += temp.data
                temp = temp.next
                if sum_value == 0:
                    self.head = current.next
                    current = self.head
                    break
            if sum_value != 0:
                current = current.next

def take_input_and_delete_zero_sum_elements():
    linked_list = LinkedList()

    n = int(input("Enter the number of elements: "))
    print("Enter the elements:")
    for _ in range(n):
        data = int(input())
        linked_list.append(data)

    print("Linked List before deleting zero-sum elements:")
    linked_list.display()

    linked_list.delete_elements_with_zero_sum()

    print("Linked List after deleting zero-sum elements:")
    linked_list.display()

if __name__ == "__main__":
    take_input_and_delete_zero_sum_elements()
```

### Q.2 Reverse a linked list in groups of given size

```python
class Node:

    # Constructor to initialize the node object
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:

    # Function to initialize head
    def __init__(self):
        self.head = None

    def reverse(self, head, k):

        if head == None:
            return None
        current = head
        next = None
        prev = None
        count = 0

        # Reverse first k nodes of the linked list
        while(current is not None and count < k):
            next = current.next
            current.next = prev
            prev = current
            current = next
            count += 1

        # next is now a pointer to (k+1)th node
        # recursively call for the list starting
        # from current. And make rest of the list as
        # next of first node
        if next is not None:
            head.next = self.reverse(next, k)

        # prev is new head of the input list
        return prev

    # Function to insert a new node at the beginning
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    # Utility function to print the linked LinkedList
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data,end=' --> ')
            temp = temp.next
        print('None')


# Driver program
llist = LinkedList()
llist.push(9)
llist.push(8)
llist.push(7)
llist.push(6)
llist.push(5)
llist.push(4)
llist.push(3)
llist.push(2)
llist.push(1)

print("Given linked list")
llist.printList()
llist.head = llist.reverse(llist.head, 3)

print ("\nReversed Linked list")
llist.printList()
```

### Q.3 Merge a linked list into another linked list at alternate positions.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

def merge_alternate_positions(list1, list2):
    merged_list = LinkedList()
    current1, current2 = list1.head, list2.head

    while current1 or current2:
        if current1:
            merged_list.append(current1.data)
            current1 = current1.next
        if current2:
            merged_list.append(current2.data)
            current2 = current2.next

    return merged_list

def take_input_and_merge_alternate_positions():
    list1 = LinkedList()
    list2 = LinkedList()

    n = int(input("Enter the number of elements for the first list: "))
    print("Enter the elements for the first list:")
    for _ in range(n):
        data = int(input())
        list1.append(data)

    m = int(input("Enter the number of elements for the second list: "))
    print("Enter the elements for the second list:")
    for _ in range(m):
        data = int(input())
        list2.append(data)

    merged_list = merge_alternate_positions(list1, list2)

    print("First List:")
    list1.display()

    print("Second List:")
    list2.display()

    print("Merged List at Alternate Positions:")
    merged_list.display()

if __name__ == "__main__":
    take_input_and_merge_alternate_positions()
```

### Q.4 In an array, Count Pairs with given sum

```python
def count_pairs_with_sum(arr, target_sum):
    count = 0
    num_dict = {}

    for num in arr:
        complement = target_sum - num
        if complement in num_dict:
            count += num_dict[complement]
        if num in num_dict:
            num_dict[num] += 1
        else:
            num_dict[num] = 1

    return count

def take_input_and_count_pairs():
    arr = []
    n = int(input("Enter the number of elements in the array: "))
    print("Enter the elements of the array:")
    for i in range(n):
        element = int(input())
        arr.append(element)

    target_sum = int(input("Enter the target sum: "))

    result = count_pairs_with_sum(arr, target_sum)

    print(f"Number of pairs with sum {target_sum} is {result}")

if __name__ == "__main__":
    take_input_and_count_pairs()
```

### Q.6 Find the Kth largest and Kth smallest number in an array

```python
class KthMinMaxFinder:
    def __init__(self, arr):
        self.arr = arr

    def find_kth_max(self, k):
        sorted_arr = sorted(self.arr, reverse=True)
        if 0 < k <= len(sorted_arr):
            return sorted_arr[k - 1]
        else:
            return None

    def find_kth_min(self, k):
        sorted_arr = sorted(self.arr)
        if 0 < k <= len(sorted_arr):
            return sorted_arr[k - 1]
        else:
            return None

if __name__ == "__main__":
    array = list(map(int, input("Enter the array: ").split(",")))
    k = int(input("Enter k: "))

    kth_min_max_finder = KthMinMaxFinder(array)

    kth_max = kth_min_max_finder.find_kth_max(k)
    kth_min = kth_min_max_finder.find_kth_min(k)

    if kth_max is not None:
        print(f"Kth maximum element is: {kth_max}")
    else:
        print("Invalid value of k for finding the maximum element.")

    if kth_min is not None:
        print(f"Kth minimum element is: {kth_min}")
    else:
        print("Invalid value of k for finding the minimum element.")
```

### Q.9 Evaluate a postfix expression using stack

```python
# Class to convert the expression
class Evaluate:

    # Constructor to initialize the class variables
    def __init__(self, capacity):
        self.top = -1
        self.capacity = capacity
        
        # This array is used a stack
        self.array = []

    # Check if the stack is empty
    def isEmpty(self):
        return True if self.top == -1 else False

    # Return the value of the top of the stack
    def peek(self):
        return self.array[-1]

    # Pop the element from the stack
    def pop(self):
        if not self.isEmpty():
            self.top -= 1
            return self.array.pop()
        else:
            return "$"

    # Push the element to the stack
    def push(self, op):
        self.top += 1
        self.array.append(op)

    # The main function that converts given infix expression
    # to postfix expression
    def evaluatePostfix(self, exp):

        # Iterate over the expression for conversion
        for i in exp:

            # If the scanned character is an operand
            # (number here) push it to the stack
            if i.isdigit():
                self.push(i)

            # If the scanned character is an operator,
            # pop two elements from stack and apply it.
            else:
                val1 = self.pop()
                val2 = self.pop()
                self.push(str(eval(val2 + i + val1)))

        return int(self.pop())



# Driver code
if __name__ == '__main__':
    exp = input("Enter the Expression : ")
    obj = Evaluate(len(exp))

    # Function call
    print("postfix evaluation: %d" % (obj.evaluatePostfix(exp)))
    
```

### Q.10 Implement a queue using the stack data structure


```python
class QueueUsingStacks:
    def __init__(self):
        self.input_stack = []
        self.output_stack = []

    def is_empty(self):
        return len(self.input_stack) == 0 and len(self.output_stack) == 0

    def enqueue(self, item):
        self.input_stack.append(item)

    def dequeue(self):
        if not self.output_stack:
            while self.input_stack:
                self.output_stack.append(self.input_stack.pop())

        if self.output_stack:
            return self.output_stack.pop()
        else:
            return None

def take_input_and_operate_queue():
    queue = QueueUsingStacks()

    while True:
        print("Queue Operations:")
        print("1. Enqueue")
        print("2. Dequeue")
        print("3. Quit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            item = input("Enter the element to enqueue: ")
            queue.enqueue(item)
            print(f"{item} enqueued.")
        elif choice == 2:
            if not queue.is_empty():
                item = queue.dequeue()
                print(f"{item} dequeued.")
            else:
                print("Queue is empty.")
        elif choice == 3:
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    take_input_and_operate_queue()
```
